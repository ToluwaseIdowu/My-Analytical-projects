```python
# Sample responses
responses = [
    "Everything, Work, started driving lessons, relationship with girlfriend still going strong.",
    "My behaviour",
    "The end of intervention I used for Daniel had some slightly different questions:",
    "Was it easy getting to appointments? Yes, most of the time because I didn't have to go anywhere.",
    "Doing more productive things",
    "working and passes theory test",
    "I am now working, earning money from my part time job, I have also finished college and received my qualifications.",
    "I am more intelligent about the court system",
    "My mental health, my temper. I understand how my emotions work more and how to manage them. It's not worth it\" \"I feel like I believe in myself more\".",
    "I've got a tutor now and I've not been in trouble for ages.",
    "Done cscs card and being more mindful of peers and behaviour",
    "Everything imaginable",
    "I`ve not got into trouble or done anything bad",
    "I have got a job now",
    "Nothing",
    "nothing really",
    "Christmas went well (family life has previously been quite difficult)",
    "Things at home are better",
    "I am going to be moving to a new area. I have kept out of trouble.",
    "Not really",
    "Same as before",
    "I haven't been in trouble for ages.",
    "Positive habits and routine. Work, income, time-keeping.",
    "Completed by GCSEs and left school",
    "Every thing is good",
    "nothing",
    "Stayed the same. I haven't been charged with any new offences so that is better.",
    "Not drinking, offending or going missing",
    "I passed my bike test",
    "Relationship with Ellie. Healthier relationship than previous relationships."
]

# Keyword lists based on your example
positive_keywords = [
    'work', 'job', 'earning', 'qualified', 'college', 'passed', 'driving', 'relationship', 'better',
    'not been in trouble', 'productive', 'income', 'time-keeping', 'cscs', 'mindful', 'healthier', 'good', 'finished', 'received', 'christmas went well', 'left school'
]
negative_keywords = [
    'nothing', 'not really', 'same as before'
]
neutral_keywords = [
    'my behaviour', 'the end of intervention', 'was it easy getting to appointments'
]

def classify_response(text):
    # Normalize text
    text_lower = text.lower()
    # Negative check first for "nothing", "not really", etc.
    for kw in negative_keywords:
        if kw in text_lower:
            return 'Negative'
    # Positive check
    for kw in positive_keywords:
        if kw in text_lower:
            return 'Positive'
    # Neutral check
    for kw in neutral_keywords:
        if kw in text_lower:
            return 'Neutral'
    # Default - if it seems positive (not negative or neutral)
    return 'Positive'

# Print results with "color codes"
for response in responses:
    category = classify_response(response)
    color = {'Positive': 'Yellow', 'Negative': 'Red', 'Neutral': 'Purple'}[category]
    print(f"{response} -> {category} ({color})")
```
