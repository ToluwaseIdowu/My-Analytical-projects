```python
#Prophet model for the time series analyses
from prophet import Prophet
from prophet.plot import add_changepoints_to_plot, plot_cross_validation_metric
from prophet.diagnostics import cross_validation, performance_metrics

#streamlit
import streamlit as st
#
#visualization
import plotly.graph_objs as go

import yfinance as yf
import pandas as pd
import numpy as np
import seaborn as sns
from datetime import date
from pandas import read_csv
from pandas import set_option
from prophet.plot import plot_plotly
from plotly import graph_objs as go
from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error,r2_score, mean_absolute_error
plt.rcParams['figure.figsize']=(20,10)
plt.style.use('ggplot')

df = pd.DataFrame()

crypto_symbols = ['BTC',]
cryptocurrencies = ['Bitcoin']

for coins in crypto_symbols:
    coin_index = crypto_symbols.index(coins)
    yf_data = yf.Ticker(f'{coins}-USD').history(start='2020-01-01', end= date.today().strftime("%Y-%m-%d"), interval = '1d')

    cdf = pd.DataFrame(yf_data)
    cdf['name'] = cryptocurrencies[coin_index]
    cdf['symbol'] = coins
    df = df.append(cdf)

#Reset index and make date a column
df.reset_index(inplace=True)
#change date to datetime format
df['Date'] = pd.to_datetime(df['Date'])

df.head()
```

```python
#Dropping irrelivant variable
new_df = df.drop(['Open','High','Low','Volume','Dividends','Stock Splits','name','symbol'], axis=1)

#Renaming date to ds and close price to y
new_df.rename(columns={'Date': 'ds', 'Close': 'y'}, inplace=True)


#Seeing the change in the names
new_df.head()



```

```python
plt.figure()
new_df.set_index('ds').y.plot().get_figure()
```

```python
#Checking Data information
new_df.info()
```

```python
#TRAIN TEST SPLIT
new_df_train = new_df[new_df['ds']<='2022-01-01']
new_df_test = new_df[new_df['ds']> '2022-01-01']

print('The training dataset has', len(new_df_train),'records, ranging from', new_df_train['ds'].min(), 'to', new_df_train['ds'].max())

print('The testing dataset has', len(new_df_test),'records, ranging from', new_df_test['ds'].min(), 'to', new_df_test['ds'].max())

```

```python
#create the prophet model with confidence interval of 95%
m = Prophet(interval_width=0.95, n_changepoints=7)

#Fit the model using the training dataset
m.fit(new_df_train)
```

```python
#Create a future dataframe for prediction
future = m.make_future_dataframe(periods= 31)

#forecast the future dataframe value
forecast = m.predict(future)

# check the forecasted values and upper/lower bound
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()

```

```python
#visualize the forecast
fig = m.plot(forecast)
ax = fig.gca()
ax.plot(new_df_test["ds"], new_df_test["y"], 'r.')
```

```python
#Time series Decomposition
# visualize the components
m.plot_components(forecast);
```

```python
# Identify Change Points
print(f'There are {len(m.changepoints)} change points. \nThe change points dates are\n {new_df.loc[new_df["ds"].isin(m.changepoints)]}')
```

```python
# Change points to plot
fig = m.plot(forecast)
a = add_changepoints_to_plot(fig.gca(), m, forecast)
```

```python
# Cross validation
new_df_cv = cross_validation(m, initial='500 days', period='60 days', horizon = '30 days', parallel="processes")
new_df_cv.head()
```
